```python
import numpy as np
from scipy.optimize import root_scalar, fsolve
import matplotlib.pyplot as plt

# Параметры модели Ланга–Кобаяши (случай Q >> 1)
alpha = 5.0
v = 10.0
Q = 10.0          
omega0 = 0.0

# Опорные параметры для первичного поиска мод
gamma_op = 8.0
h_op = 8.0
```

```python
C = gamma_op * h_op * np.sqrt(1 + alpha**2)
omega0_h = omega0 * h_op

eta_min = omega0_h - C - 2*np.pi
eta_max = omega0_h + C + 2*np.pi
num_points = max(5000, int(10 * C))
etas = np.linspace(eta_min, eta_max, num_points)

def F(eta):
    return eta - omega0_h + C * np.sin(eta + np.arctan(alpha))

F_vals = F(etas)
roots = []
for i in range(len(etas)-1):
    if F_vals[i] * F_vals[i+1] < 0:
        try:
            sol = root_scalar(F, bracket=[etas[i], etas[i+1]], method='brentq')
            if not any(abs(sol.root - r) 
                       
                       < 1e-6 for r in roots):
                roots.append(sol.root)
        except:
            pass
roots = np.array(sorted(roots))
print(f"Найдено математических корней: {len(roots)}")
```
